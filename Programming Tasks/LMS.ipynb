```python
import pandas as pd
```

```python
df = pd.read_csv('random3.csv', names = ['x1','x2','y'])
df['x0'] =1
```

```python
df.head()
```

```python
def LMS(eta,threshold):
    thresholdreached = True
    w0, w1, w2 = 0, 0, 0
    iterations = 0
    while thresholdreached == True:
        sse = 0
        iterations +=1
        grad0, grad1, grad2 = 0, 0, 0
        for i in range(len(df)):
            yhat =  w0*df['x0'][i] + w1*df['x1'][i] + w2*df['x2'][i]
            error = float(df['y'][i] - yhat)
            sse += float(error)**2
            grad0 += error*df['x0'][i]
            grad1 += error*df['x1'][i]
            grad2 += error*df['x2'][i]
        
        w0 = w0 + float(eta*grad0)
        w1 = w1 + float(eta*grad1)
        w2 = w2 + float(eta*grad2)
        print(iterations,w0, w1, w2, sse)
        if float(sse)<=threshold:
            thresholdreached = False
            break
```

```python
LMS(eta=0.00005, threshold=0.001)
```
