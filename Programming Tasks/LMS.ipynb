```python
import pandas as pd
import sys
```

```python
df = pd.read_csv(str('random3.csv'), names = ['x1','x2','y'])
```

```python
df.head()
```

```python
def LMS(eta,threshold):
    thresholdreached = True
    w0, w1, w2 = 0, 0, 0
    iterations = 0
    prev_sse = 0
    while thresholdreached == True:
        sse = 0
        grad0, grad1, grad2 = 0, 0, 0
        for i in range(len(df)):
            yhat =  w0*1 + w1*df['x1'][i] + w2*df['x2'][i]
            error = float(df['y'][i] - yhat)
            sse += float(error)**2
            grad0 += error*1
            grad1 += error*df['x1'][i]
            grad2 += error*df['x2'][i]
                
        w0 = w0 + float(eta*grad0)
        w1 = w1 + float(eta*grad1)
        w2 = w2 + float(eta*grad2)
        
        print(iterations,w0, w1, w2, sse)
        
        iterations +=1
        
        if abs(sse-prev_sse)<threshold:
            thresholdreached = False
            break
        
        prev_sse = sse
        
        

        
```

```python
iterations = LMS(eta=0.00005, threshold=0.0001)
```
